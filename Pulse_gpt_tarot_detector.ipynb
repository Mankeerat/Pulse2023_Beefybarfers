```python
import requests
import numpy as np
import urllib.parse
import os
import openai
import json
import speech_recognition as sr
import keyboard
from gtts import gTTS
from IPython.display import Audio, display
```

```python
dset = open('cards.json')
data = json.load(dset)
```

```python
language = 'en'

inque = "What is in your mind?"

inobj = gTTS(text=inque, lang=language, slow=False)
inobj.save("inque.mp3")
audio = Audio("inque.mp3", autoplay=True)
display(audio)

r = sr.Recognizer()

with sr.Microphone() as source:
    print("Talk")
    r.adjust_for_ambient_noise(source)
    audio_text = r.listen(source)
    print("Time over, thanks")
try:
    text = r.recognize_google(audio_text)
    print("Text: "+r.recognize_google(audio_text))
except:
    print("Sorry, I did not get that")

question = text

cardpic = "Choose 3 numbers between 0 and 77, make sure the numbers are unique from one another(for example 31 23 44)\n"

hmobj = gTTS(text=cardpic, lang=language, slow=False)
hmobj.save("hmque.mp3")
audio2 = Audio("hmque.mp3", autoplay=True)
display(audio2)

cards = []

for i in range(3):
    cardidx = input("type")
    cards.append(data[i])

send_data = "I want you to act as an amazing tarot card reader and a therapist who is also a person who can understand JSON format. I will provide you my current situation and related question. Also, I will provide you three formatted JSON strings that represent the cards I drew. The first card represents the past, the second represents the present, and the third represents the future. They will be seperated with the curley braces. Each JSON string will provide you fortune telling, keywords, meanings, shadow, name, rank, and suit of the said cards. Your job is to analyze the cards and provide some answers to my question based on the cards. Make sure to directly refer to the cards I gave you before giving your thoughts on each cards."+"\nHere is my question: "+question+"\nand here are the cards that I drew: "+"\n"+str(cards[0])+"\n"+str(cards[1])+"\n"+str(cards[2])+ "\n"

mid = "Hmm, let's see what your fortune holds for you"
midobj = gTTS(text=mid, lang=language, slow=False)
midobj.save("mid.mp3")
audio = Audio("mid.mp3", autoplay=True)
display(audio)

API_KEY = "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"
openai.api_key = API_KEY

model = 'text-davinci-003'

response = openai.Completion.create(
    prompt = send_data,
    model = model, 
    max_tokens = 1000, 
    temperature = 0.9
)

hehe = json.loads(str(response))
myanswertext = hehe['choices'][0]['text']
myobj = gTTS(text=myanswertext, lang=language, slow=False)
myobj.save("welcome.mp3")

Audio("welcome.mp3", autoplay=True)
```
